```python
#Conversion de la colonne (avec l'argument errors=coerce qui permet de convertir automatiquement les lignes qui ne sont pas des nombres en NaN)
#Puis remplacement des NaN en 0
sous_nutrition['Valeur'] = pd.to_numeric(sous_nutrition['Valeur'], errors='coerce')
sous_nutrition=sous_nutrition.fillna(0)
```

```python
#changement du nom de la colonne Valeur par sous_nutrition
sous_nutrition = sous_nutrition.rename(columns={'Valeur':'Sous_nutrition'});
```

```python
#Multiplication de la colonne sous_nutrition par 1000000
sous_nutrition['Sous_nutrition']=sous_nutrition['Sous_nutrition']*1E6
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.1 - Proportion de personnes en sous nutrition</h3>
</div>

```python
# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017
# Extract the average year from the 'Année' column in sous_nutrition
sous_nutrition['Année'] = sous_nutrition['Année'].str.split('-').apply(lambda x: (int(x[0]) + int(x[1])) / 2 if isinstance(x, list) and len(x) == 2 else None)

# Convert the 'Année' column to integer type
sous_nutrition['Année'] = sous_nutrition['Année'].astype(int)

#jointure entre les tables population et sous_nutrition en ciblant l'année selon la zone et l'année
jointure_population_sous_nutrition=pd.merge(population,sous_nutrition,how='inner',on=['Année','Zone'])
jointure_population_sous_nutrition=jointure_population_sous_nutrition[jointure_population_sous_nutrition['Année']==2017]
jointure_population_sous_nutrition=jointure_population_sous_nutrition.fillna(0)
```

```python
#Affichage du dataset
jointure_population_sous_nutrition.head()
```

```python
#Calcul et affichage du nombre de personnes en état de sous nutrition
nombre_personnes_sous_nutrition=jointure_population_sous_nutrition['Sous_nutrition'].sum()
population_totale=jointure_population_sous_nutrition['Population'].sum()
proportion_personnes_sous_nutrition=(nombre_personnes_sous_nutrition/population_totale)*100
print("La proportion de personnes en situation de sous_nutrition est de {} %".format(round(proportion_personnes_sous_nutrition)))
```

<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Nombre théorique de personne qui pourrait être nourries</h3>

```python
#Combien mange en moyenne un être humain ? Source => https://www.vidal.fr/sante/nutrition/equilibre-alimentaire-adulte/recommandations-nutritionnelles-adulte.html

#2000 en moyenne pour les femmes, 2500 en moyenne pour les hommes => 2250 kcal par jour en moyenne
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
jointure_population_dispo_alimentaire=pd.merge(population[population['Année']==2017],dispo_alimentaire,how='inner',on=['Zone'])
```

```python
#Affichage du nouveau dataframe
jointure_population_dispo_alimentaire.head()
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
jointure_population_dispo_alimentaire['dispo_kcal']=jointure_population_dispo_alimentaire['Disponibilité alimentaire (Kcal/personne/jour)']*jointure_population_dispo_alimentaire['Population']
```

```python
#Calcul du nombre d'humains pouvant être nourris
jointure_population_dispo_alimentaire['Nombre_humains_nourris']=jointure_population_dispo_alimentaire['dispo_kcal']/2250
print("En 2017, nous aurions pu nourrir {} humains".format(round(jointure_population_dispo_alimentaire['Nombre_humains_nourris'].sum())))
```

```python
#création d'un dataframe avec uniquement la Thaïlande

#filtrage des donnees
donnees_thailande_sous_nutrition=jointure_population_sous_nutrition[jointure_population_sous_nutrition['Zone']=='Thaïlande']
donnees_thailande_dispo_alimentaire=jointure_population_dispo_alimentaire[jointure_population_dispo_alimentaire['Zone']=='Thaïlande']

#jointure
donnees_thailande=pd.merge(donnees_thailande_sous_nutrition,donnees_thailande_dispo_alimentaire,how='inner',on=['Zone','Année'])

#suppression du doublon de colonne population
donnees_thailande=donnees_thailande.drop('Population_y',axis=1)
donnees_thailande=donnees_thailande.rename(columns={'Population_x':'Population'})

#affichage
donnees_thailande.head()
```

```python
#Calcul de la sous nutrition en Thaïlande

sous_nutrition_thailande=donnees_thailande_sous_nutrition['Sous_nutrition'].sum()

print("La Thaïlande comporte {} habitants dans une situation de sous nutrition, soit {} % de sa population totale".format(round(sous_nutrition_thailande), round(sous_nutrition_thailande/donnees_thailande_sous_nutrition['Population'].sum()*100)))

#Calcul de la disponibilité pour la Thaïlande

dispo_par_habitant_thailande=donnees_thailande['Disponibilité alimentaire (Kcal/personne/jour)'].sum()
print("La Thaïlande a une disponibilité alimentaire de {} kcal par habitant".format(round(dispo_par_habitant_thailande)))
```

```python
#filtre sur le manioc
donnees_thailande_manioc=donnees_thailande[donnees_thailande['Produit']=='Manioc']
display(donnees_thailande_manioc)
```

```python
#détails de la disponibilité intérieure pour le manioc

for element in ['Aliments pour animaux','Pertes','Nourriture','Autres Utilisations','Semences','Traitement'] :
  print(element, ":", round(donnees_thailande_manioc[element].sum()), "kgs, soit", round((donnees_thailande_manioc[element].sum()/donnees_thailande_manioc['Disponibilité intérieure'].sum())*100), "% de la disponibilité totale de manioc")

for element in ['Exportations - Quantité','Importations - Quantité','Production'] :
  print(element, ":", round(donnees_thailande_manioc[element].sum()), "kgs, soit", round((donnees_thailande_manioc[element].sum()/donnees_thailande_manioc['Production'].sum())*100), "% de la production de manioc")

for element in ['Nombre_humains_nourris'] :
  print(element, ":", round(donnees_thailande_manioc[element].sum()), "soit", round((donnees_thailande_manioc[element].sum()/donnees_thailande_manioc['Population'].sum())*100), "% de sa population")
```
